# Panda Visualizing

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/enricokoegler/Desktop/Data Analysis Bootcamp/Python/Bundesliga/bundesliga.csv',  index_col=0)
df
```

```python
ax = df[['Diff']].plot(kind='line', title='S-U-N', xlabel='Platzierung', ylabel='Anzahl')
ax = ax.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.style.use('classic')
plt.show()
```

```python
ax = df.plot.scatter(0, 'G')
ax.set_xticklabels(df['Team'], rotation=90)
plt.style.use('classic')
plt.show()
```

```python
df.sort_values('Platz', ascending=False, inplace=True)
ax = df[['S', 'U', 'N']].plot(kind='barh', stacked=True, color=['green', 'orange', 'red'])
ax.set_yticklabels(df['Team'])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
df.sort_values('Platz', ascending=True, inplace=True)
plt.style.use('classic')
plt.show()
```

```python
df2 = df.copy()
df2.sort_values('Platz', ascending=False, inplace=True)
df2['S'] = df2['S'] * 3
ax = df2[['S', 'U']].plot(kind='barh', stacked=True, color=['green', 'orange'])
ax.set_yticklabels(df2['Team'])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
for i in range(5, df2['Punkte'].max(), 5):
    ax.axvline(x=i, color='gray', linestyle='--', linewidth=0.5)
plt.style.use('dark_background')
plt.show()
```
