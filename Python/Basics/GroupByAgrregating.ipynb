# GroupBy and Aggregating

```python
import pandas as pd
df = pd.read_excel('/Users/enricokoegler/Library/Mobile Documents/com~apple~CloudDocs/Privat/Android Apps/WBPA Material /Datenbank.xlsx', sheet_name='Karten')
df = df.iloc[:, 0:13]
df = df.drop(['Average', 'Difference', 'SetId'], axis=1)
df
```

```python
df[df['Set']=='Award']
```

```python
gr = df.groupby('Set')
headers = df.columns[1:9]
for h in headers:
    x = gr.mean(numeric_only=True).round(1).sort_values(by=h, ascending=False)[[h]]
    print(x)
    #gr[h] = range(1, len(gr) + 1)
```

```python
gr = df.groupby('Set')
gr.count()[['Name']].sort_values('Name', ascending=False).rename(columns={'Name': 'Anzahl'})
```

```python
gr = df.drop('Name', axis=1).groupby('Set')
gr.median().sort_values(by='FIN', ascending=False).round(0)
```

```python
df.drop('Name', axis=1).groupby('Set').agg({'FIN':['mean', 'max','median'], 'DEF':['mean', 'median']}).round()
```
