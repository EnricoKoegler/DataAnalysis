# Merging 

```python
import pandas as pd
from fuzzywuzzy import process

df1 = pd.read_csv('/Users/enricokoegler/Desktop/Data Analysis Bootcamp/Python/Bundesliga/bundesliga.csv', index_col=0)
df2 = pd.read_csv('/Users/enricokoegler/Desktop/Data Analysis Bootcamp/Python/Bundesliga/bundesliga_gehalt.csv', index_col=0)
df1
```

```python
def find_best_match(name, choices):
    return process.extractOne(name, choices)[0]

# DataFrames nach Namen zuordnen
df1['MappedName'] = df1['Team'].apply(lambda x: find_best_match(x, df2['Verein']))
merged_df = pd.merge(df1, df2, left_on='MappedName', right_on='Verein', how='inner')
merged_df.drop(['MappedName', 'Verein', 'Schätzwert in %'], axis=1, inplace=True)
merged_df
```

```python
df1 = df1[['Team', 'G']]
merged_df = df1.merge(df1, left_on='G', right_on='G')
merged_df = merged_df[merged_df['Team_x'] != merged_df['Team_y']]
merged_df
```
